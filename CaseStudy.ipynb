```python
def bar_plot(variable) :
  var = df[variable]
  varValue = var.value_counts()
  plt.figure(figsize = (39, 33))
  plt.bar(varValue.index, varValue)
  plt.xticks(varValue.index, varValue.index.values)
  plt.ylabel("Frequency")
  plt.title(variable)
  plt.show()
  print("{}: \n {}".format(variable, varValue))
```

```python
columns_plot = ['artist_count', 'released_year','released_month', 'released_day', 'in_spotify_playlists','in_spotify_charts', 'streams', 'in_apple_playlists', 'in_apple_charts','in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm']
for i in columns_plot:
  bar_plot(i)
```

```python
fig, axes = plt.subplots(5, 2, figsize=(12, 25))
for ax, col in zip(axes.flat, df.columns):
    sns.histplot(df[col], bins=30, kde=True, ax=ax)
    ax.set_title(col)
plt.tight_layout()
plt.show()
```

```python
for col in df.columns:
    value_counts = df[col].value_counts()
    threshold = 0.02 * sum(value_counts)
    value_counts_filtered = value_counts[value_counts > threshold]
    others = sum(value_counts[value_counts <= threshold])
    if others > 0:
        value_counts_filtered["Others"] = others
    plt.figure(figsize=(7, 7))
    value_counts_filtered.plot.pie(autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})
    plt.title(col)
    plt.ylabel('')
    plt.show()
```

```python
df = pd.DataFrame({
    'released_year': np.tile(np.arange(1, 11), 4),
    'artist_count': np.random.rand(40),
    'in_spotify_playlists': np.repeat(['parietal', 'frontal'], 20),
    'streams': np.tile(['stim', 'cue'], 20)
})

# lmplot in requested format
sns.lmplot(
    x='released_year',
    y='artist_count',
    hue='in_spotify_playlists',
    col='streams',  # Splitting into separate plots for different 'event' values
    data=df,
    aspect=1.2,  # Adjusting aspect ratio
    height=5,  # Adjusting figure height
    markers=['o', 's'],  # Custom markers for clarity
    scatter_kws={'alpha': 0.8}  # Adjusting scatter transparency
)

plt.show()
```
